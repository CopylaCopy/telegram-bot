```python
import pandas as pd
import re
import numpy as np
pd.set_option('display.max_rows', 500)
```

```python
def initials(text):
        t = re.findall(r'([А-ЯЁ])[., ]*([А-ЯЁ])', text)
        return ''.join(t[0])
```

```python
def standart_spot(spot):
    if spot.startswith('.'):
        spot = spot[1:]
    if 'ИП' in spot:
        t = re.findall(r'([ИП]{2})?[ ]?([А-ЯЁёа-я]+) ([А-ЯЁ][., ]+[А-ЯЁ][.,]*)[ ]?([ИП]{2})?(.*)', spot)
        if t:
            text = list(t[0])
            if 'ИП' in text:
                ind = text.index('ИП')
                text.pop(ind)
            if '' in text:
                ind = text.index('')
                text.pop(ind)
            init = initials(text[1])
            print([f'{text[0]} {init}', text[2], 'ИП'])
            return [f'{text[0]} {init}', text[2], 'ИП']
            
        t = re.findall(r'([ИП]{2})?[ ]?([А-ЯЁёа-я]+)[ ]?([ИП]{2})?(.*)', spot)
        if t:
            text = list(t[0])
            if 'ИП' in text:
                ind = text.index('ИП')
                text.pop(ind)
            if '' in text:
                ind = text.index('')
                text.pop(ind)
            print([f'{text[0]}', text[1], 'ИП'])
            return [f'{text[0]}', text[1], 'ИП']
    if 'ООО' in spot:
        spot = spot.replace('\\', '')
        t = re.findall(r'^[\"]*(([О]{3})[ ]?[.А-Яа-яёЁ№0-9\\\"\-\+ ]+)(.*)', spot)
        if t:
            text = list(t[0])
            
            if 'ООО' in text:
                ind = text.index('ООО')
                text.pop(ind)
            print([text[0], text[1], 'ООО'])
            return [text[0], text[1], 'ООО']
        t = re.findall(r'^([.А-Яа-яёЁ№0-9\\\\"\-\+ ]+) [.]?([О]{3})[ ]?[.\\\'\"]*(.*)', spot)
        if t:
            text = list(t[0])
            if 'ООО' in text:
                ind = text.index('ООО')
                text.pop(ind)
            print([text[0], text[1], 'ООО'])
            return [text[0], text[1], 'ООО']
        else:
            return [spot, '', '']
    else:
        return [spot.replace('.', ''), '', '']
                
                    
    
```

```python
outlets['standart'] = outlets['Торг_точка_грязная'].apply(standart_spot)
```
